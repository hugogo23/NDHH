```python
import matplotlib.pyplot as plt
import os 
import json
```

```python
def sort_list(x_y):
    x_y = [ [int(x),y] for x,y in x_y ]
    #x_y = [ [x,y] for x,y in x_y if x % 10 == 0]
    x_y.sort()
    x,y = zip(*x_y)
    return x, y
    
def plot_coherence(json_fp):
    data = json.loads(open(json_fp,"r").read() )
    
    words5_x, words5_y = sort_list( list( data["5words"].items() ) )
    words10_x, words10_y = sort_list( list( data["10words"].items() ) )
    
    plt.plot( words5_x, words5_y, color='r', marker='o',linestyle='dashed', label="top5 word coherence")
    plt.plot( words10_x, words10_y, color='b', marker='*',linestyle='dashed', label="top10 word coherence")
    if "20words" in data:
        words20_x, words20_y = sort_list( list( data["20words"].items() ) )
        plt.plot( words20_x, words20_y, color='g', marker='.',linestyle='dashed', label="top20 word coherence")
    
    plt.legend()
    plt.show()
```

```python
file_folder = "/home/linyu.linyu/health.research/health/data/cls_output/topic_out/"
file1 = file_folder + "evaluate_num_topic_feature1w.json"
plot_coherence(file1)


```

```python
## how to use plt see :https://zhuanlan.zhihu.com/p/109245779
#1 基础绘图
x=[1, 2, 3, 4]
y=[1, 4, 9, 16]
plt.plot(x,y)
plt.show()


#2 定义绘图属性
'''
color：线条颜色，值r表示红色（red）
marker：点的形状，值o表示点为圆圈标记（circle marker）
linestyle：线条的形状，值dashed表示用虚线连接各点
'''
plt.plot(x, y, color='r',marker='o',linestyle='dashed')
#plt.plot(x, y, 'ro')
'''
axis：坐标轴范围
语法为axis[xmin, xmax, ymin, ymax]，
也就是axis[x轴最小值, x轴最大值, y轴最小值, y轴最大值]
'''
plt.axis([0, 6, 0, 20])
plt.show()


#3 给图片提阿甲注释和标题等
x=[1, 2, 3, 4]
y=[1, 4, 9, 16]

#第2步：使用plot绘制线条 第1个参数是x的坐标值，第2个参数是y的坐标值
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
plt.annotate('write_here', xy=(2,5), xytext=(2, 10),
            arrowprops=dict(facecolor='black', shrink=0.01),
            )

#第3步：显示图形
plt.show()
```
